```python
import numpy as np
from Lineshape import *
import matplotlib.pyplot as plt
```

```python
bin_to_freq = 0.0015287  # MHz per bin
start_freq = 212.6  # Starting frequency in MHz

# Create an independent variable (frequency in MHz) and extract the relevant data
x_full_bins = np.arange(500)  # Full range of bins
x = start_freq + x_full_bins * bin_to_freq  # Convert bins to frequency
U = 2.4283e1
Cknob = .0647
cable = 22/2
eta = 1.04e-2
phi = 6.1319
Cstray = 10**(-15)
shift = -2.0464e-2
sig = .01
gam = .01
center = 213
A = .005
Sig = Baseline(x,U,Cknob,eta,cable,Cstray,phi,shift)
v = Voigt(x,A,sig,gam,center)
```

```python
plt.figure()
# plt.plot(x,Sig,label = 'Baseline')
# plt.plot(x,v, label = 'Voigt')
plt.plot(x,Sig+v, label = 'Baseline + Voigt')
plt.legend()

```
